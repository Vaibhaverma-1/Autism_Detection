## üìä Data Description: Autism Screening Dataset

This dataset contains responses from an **Autism Spectrum Disorder (ASD) screening test**, including demographic information and screening question answers.

### üîπ Features & Explanation

| Feature | Type | Description |
|---------|------|-------------|
| **A1 (Q1)** | Binary (0,1) | **Does your child look at you when you call his/her name?** |
| **A2 (Q2)** | Binary (0,1) | **How easy is it for you to get eye contact with your child?** |
| **A3 (Q3)** | Binary (0,1) | **Does your child point to indicate that they want something?** |
| **A4 (Q4)** | Binary (0,1) | **Does your child point to share interest with you?** |
| **A5 (Q5)** | Binary (0,1) | **Does your child pretend?** |
| **A6 (Q6)** | Binary (0,1) | **Does your child follow where you‚Äôre looking?** |
| **A7 (Q7)** | Binary (0,1) | **If someone in the family is visibly upset, does your child try to comfort them?** |
| **A8 (Q8)** | Binary (0,1) | **Would you describe your child‚Äôs first words as clear and meaningful?** |
| **A9 (Q9)** | Binary (0,1) | **Does your child use simple gestures?** |
| **A10 (Q10)** | Binary (0,1) | **Does your child stare at nothing with no apparent purpose?** |
| **Age** | Number | **Age of the participant (in months)** |
| **Sex** | Categorical | **Gender of the participant (Male/Female)** |
| **Jaundice** | Boolean (Yes/No) | **Whether the participant was born with jaundice** |
| **Family_ASD** | Boolean (Yes/No) | **If an immediate family member has ASD** |
| **Class** | Binary (Yes/No) | **Target Variable ‚Äì ASD Traits Present or Not** |

---

### üìå Key Notes
- **A1-A10** represent responses to standardized ASD screening questions.
- **Class** is the **target variable** (Yes = ASD Traits, No = No ASD Traits).
- This dataset is useful for **predictive modeling** and **ASD screening insights**.

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb

# Initialize list to store model performance results
model_performance = []

# üìå List of models without class_weight="balanced" for those that don't need it
models = [
    (LogisticRegression(random_state=42), "Logistic Regression"),
    (SVC(probability=True, random_state=42), "Support Vector Machine"),
    (RandomForestClassifier(random_state=42), "Random Forest"),
    (GradientBoostingClassifier(random_state=42), "Gradient Boosting"),  # No class_weight in boosting
    (DecisionTreeClassifier(random_state=42), "Decision Tree"),
    (KNeighborsClassifier(), "K-Nearest Neighbors"),  # No class_weight for KNN
    (GaussianNB(), "Naive Bayes")  # No class_weight for Naive Bayes
]
```

```python
# Define the evaluate_model function
def evaluate_model(model, model_name, X, y, cv=5):
    # Use cross-validation for accuracy, precision, recall, and f1 score
    accuracy_scores = cross_val_score(model, X, y, cv=cv, scoring="accuracy")
    precision_scores = cross_val_score(model, X, y, cv=cv, scoring=make_scorer(precision_score))
    recall_scores = cross_val_score(model, X, y, cv=cv, scoring=make_scorer(recall_score))
    f1_scores = cross_val_score(model, X, y, cv=cv, scoring=make_scorer(f1_score))

    # Compute average score for each metric
    scores = {
        "Model": model_name,
        "Accuracy": accuracy_scores.mean(),
        "Precision": precision_scores.mean(),
        "Recall": recall_scores.mean(),
        "F1 Score": f1_scores.mean(),
    }

    # Store results
    model_performance.append(scores)  # Append scores to the list
```

```python
# üìå List of models without class_weight="balanced" for those that don't need it
models = [
    (LogisticRegression(max_iter=5000,random_state=42), "Logistic Regression"),
    (SVC(probability=True, random_state=42), "Support Vector Machine"),
    (RandomForestClassifier(random_state=42), "Random Forest"),
    (GradientBoostingClassifier(random_state=42), "Gradient Boosting"),  # No class_weight in boosting
    (DecisionTreeClassifier(random_state=42), "Decision Tree"),
    (KNeighborsClassifier(), "K-Nearest Neighbors"),  # No class_weight for KNN
    (GaussianNB(), "Naive Bayes"),  # No class_weight for Naive Bayes
    (xgb.XGBClassifier(random_state=42), "XGBoost")  # Adding XGBoost model
]

from sklearn.model_selection import cross_val_score

# Evaluate each model using cross-validation (5-fold)
for model, name in models:
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring="accuracy")  # 5-Fold CV
    print(f"{name}: {scores.mean():.4f} ¬± {scores.std():.4f}")  # Show mean ¬± std dev
```

```python
model_performance# Evaluate each model using cross-validation (5-fold)
for model, name in models:
    evaluate_model(model, name, X_train, y_train, cv=5)  # Use the evaluate_model function

# After running the loop, print the model performance results
for performance in model_performance:
    print(f"Model: {performance['Model']}")
    print(f"Accuracy: {performance['Accuracy']:.4f}")
    print(f"Precision: {performance['Precision']:.4f}")
    print(f"Recall: {performance['Recall']:.4f}")
    print(f"F1 Score: {performance['F1 Score']:.4f}\n")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# üé® Set figure size for the subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# üåà Define Lighter Colors for each metric
colors = ["#76a5af", "#ffcc99", "#a4c2f4", "#b6d7a8"]  # Soft Blue, Light Orange, Sky Blue, Light Green

# üìä Define Metrics
metrics = ["Accuracy", "Precision", "Recall", "F1 Score"]

# üî¢ X-axis positions for each model
x_labels = [result["Model"] for result in model_performance]
x = np.arange(len(x_labels))

# üìâ Plot Histogram Bars for Each Metric in subplots
for idx, metric in enumerate(metrics):
    scores = [result[metric] for result in model_performance]

    # Calculate the average score for the current metric
    avg_score = np.mean(scores)

    # Select the subplot for the current metric
    ax = axes[idx // 2, idx % 2]  # This gets the correct position in a 2x2 grid

    # Plot bars
    ax.bar(x, scores, color=colors[idx], alpha=0.85, edgecolor="black")

    # Add a threshold line at the average score (red dotted line)
    ax.axhline(y=avg_score, color="red", linestyle="dotted", linewidth=2, label=f"Avg {metric} ({avg_score:.2f})")

    # Add titles, labels, and formatting
    ax.set_title(f"{metric} Comparison", fontsize=14, fontweight="bold", color="black")
    ax.set_ylabel("Score", fontsize=12, fontweight="bold", color="black")
    ax.set_xticks(x)
    ax.set_xticklabels(x_labels, rotation=90, ha="center", fontsize=10, fontweight="bold", color="black")
    ax.tick_params(axis='y', labelsize=10)

    # Add text for average score next to the dotted line
    ax.text(len(x)-1, avg_score + 0.01, f'{avg_score:.2f}', ha='right', fontsize=10, fontweight='bold', color='red')

# üè∑Ô∏è Formatting the overall chart layout
plt.tight_layout()

# üéØ Show the Plot
plt.show()
```

```python
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import RFE
from sklearn.utils import class_weight

# Step 1: Split the data
X_train, X_test, y_train, y_test = train_test_split(X_cleaned, y, test_size=0.2, random_state=42)

# Step 2: Scale features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 3: Feature Selection using RFE
selector = RFE(LogisticRegression(), n_features_to_select=10)  # Select top 10 features
X_train_selected = selector.fit_transform(X_train_scaled, y_train)
X_test_selected = selector.transform(X_test_scaled)

# Step 4: Class Imbalance Handling (Optional)
# Calculate class weights if class imbalance exists
class_weights = class_weight.compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weight_dict = dict(zip(np.unique(y_train), class_weights))

# Step 5: Grid Search for Hyperparameter Tuning
param_grid = {
    'C': [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000],
    'penalty': ['l2'],
    'solver': ['newton-cg', 'lbfgs', 'liblinear'],
    'max_iter': [100, 200, 300, 500],
    'class_weight': [None, 'balanced']
}

# Initialize GridSearchCV
grid_search = GridSearchCV(LogisticRegression(), param_grid, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(X_train_selected, y_train)

# Step 6: Randomized Search for more efficient hyperparameter tuning (Optional)
random_search = RandomizedSearchCV(LogisticRegression(), param_grid, n_iter=100, cv=5, n_jobs=-1, random_state=42)
random_search.fit(X_train_selected, y_train)

# Step 7: Evaluate Baseline Model
baseline_model = joblib.load('baseline_logistic_regression_model.pkl')
baseline_model.fit(X_train_selected, y_train)
baseline_y_pred = baseline_model.predict(X_test_selected)
baseline_accuracy = accuracy_score(y_test, baseline_y_pred)

# Step 8: Evaluate Tuned Model
tuned_model = grid_search.best_estimator_  # Or use random_search.best_estimator_
tuned_y_pred = tuned_model.predict(X_test_selected)
tuned_accuracy = accuracy_score(y_test, tuned_y_pred)

# Step 9: Accuracy Difference
accuracy_diff = tuned_accuracy - baseline_accuracy

# Step 10: Plot Comparison using Seaborn
accuracy_comparison = {
    'Baseline Model': baseline_accuracy,
    'Tuned Model': tuned_accuracy
}

plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
barplot = sns.barplot(x=list(accuracy_comparison.keys()), y=list(accuracy_comparison.values()), palette='muted')

# Add accuracy values on top of the bars
for p in barplot.patches:
    barplot.annotate(f'{p.get_height():.4f}',
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center',
                     fontsize=12, color='black', fontweight='bold',
                     xytext=(0, 5), textcoords='offset points')

# Add dotted line to show the difference
plt.axhline(y=baseline_accuracy, color='red', linestyle='--', label="Baseline Accuracy")

# Title and labels
plt.title(f'Comparison of Baseline and Tuned Logistic Regression Model (Accuracy Difference: {accuracy_diff:.4f})', fontsize=16, fontweight='bold')
plt.ylabel('Accuracy', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Display accuracy difference
plt.figtext(0.5, 0.05, f'Accuracy increase/decrease: {accuracy_diff:.4f}', ha='center', fontsize=12, color='black')

# Show the plot
plt.tight_layout()
plt.legend()
plt.show()

# Step 11: Print Accuracy Comparison
print(f"Baseline Model Accuracy: {baseline_accuracy:.4f}")
print(f"Tuned Model Accuracy: {tuned_accuracy:.4f}")
print(f"Accuracy Difference: {accuracy_diff:.4f}")

```
